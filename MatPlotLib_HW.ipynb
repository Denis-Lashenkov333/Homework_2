```python
#задача №1(Исследование функции)
def func(x,alpha,beta):
    try:
        return((x**beta+alpha**beta)/(x**beta))
    except:
         pass
fig = plt.figure(figsize = (20,20))#задаем размеры графика
plt.subplot(1, 2, 1)#задаем первую область
x =np.linspace(-10,10,100)#  
plt.plot(x,func(x,1,1),'--r',label='alpha=1,beta=1')
plt.plot(x,func(x,2,1),'--y',label='alpha=2,beta=1')
plt.plot(x,func(x,1,2),'--b',label='alpha=1,beta=2')
plt.plot(x,func(x,2,2),'--g',label='alpha=2,beta=2')
plt.title("ГРАФИК №1")
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
plt.xlabel("x")        
plt.ylabel("y")
plt.subplot(1, 2, 2)
plt.plot(x,func(x,1,0.5),'--r',label='alpha=1,beta=0.5')
plt.plot(x,func(x,1,-0),'--y',label='alpha=1,beta=-0')
plt.plot(x,func(x,1,-1.5),'--b',label='alpha=1,beta=-1.5')
plt.title("ГРАФИК №2")
plt.xlabel("x")        
plt.ylabel("y")
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig.savefig('p.png')

```

```python
#3(3D графики)
from mpl_toolkits import mplot3d
```

```python
#A
def f(x,y):
    return(np.cos(x**2)+np.sin(y**2))
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(60, 35) 
```
